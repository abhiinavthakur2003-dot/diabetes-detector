Features Selection :
=
---

```python
data=df[['age','gender','smoking_status','sleep_hours_per_day','screen_time_hours_per_day','family_history_diabetes','hypertension_history','cardiovascular_history','bmi','waist_to_hip_ratio','systolic_bp','diastolic_bp','heart_rate','cholesterol_total','hdl_cholesterol','ldl_cholesterol','triglycerides','glucose_fasting','glucose_postprandial','insulin_level','hba1c','diagnosed_diabetes']].copy()
```

---
Splitting Data with `Stratified Shuffle Split` so that the important categorical values ratio of the target column  remain balanced in both training and testing data :
=
---

```python
from sklearn.model_selection import StratifiedShuffleSplit
sss=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in sss.split(data,data['diagnosed_diabetes']):
    training_data=data.iloc[train_index]
    testing_data=data.iloc[test_index]
```

```python
print(training_data.shape)
print(testing_data.shape)
```

```python
training_data.head()
```

```python
training_data['smoking_status'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
train_feature_cat=training_data[['gender','smoking_status']]
test_feature_cat=testing_data[['gender','smoking_status']]
```

```python
enc=OneHotEncoder(sparse_output=False)
arr_train=enc.fit_transform(train_feature_cat)
arr_test=enc.transform(test_feature_cat)
```

```python
enctrain=pd.DataFrame(arr_train,columns=enc.get_feature_names_out(),index=train_feature_cat.index)
enctest=pd.DataFrame(arr_test,columns=enc.get_feature_names_out(),index=test_feature_cat.index)
```

```python
enctrain.shape
```

```python
enctest.shape
```

```python
enctrain.head()
```

```python
training_data.head()
```

```python
traindata=training_data[['age','sleep_hours_per_day','screen_time_hours_per_day','family_history_diabetes','hypertension_history','cardiovascular_history','bmi','waist_to_hip_ratio','systolic_bp','diastolic_bp','heart_rate','cholesterol_total','hdl_cholesterol','ldl_cholesterol','triglycerides','glucose_fasting','glucose_postprandial','insulin_level','hba1c','diagnosed_diabetes']].copy()
testdata=testing_data[['age','sleep_hours_per_day','screen_time_hours_per_day','family_history_diabetes','hypertension_history','cardiovascular_history','bmi','waist_to_hip_ratio','systolic_bp','diastolic_bp','heart_rate','cholesterol_total','hdl_cholesterol','ldl_cholesterol','triglycerides','glucose_fasting','glucose_postprandial','insulin_level','hba1c','diagnosed_diabetes']].copy()
```

```python
traindata.head()
```

```python
final_train=pd.concat([traindata,enctrain],axis=1)
final_test=pd.concat([testdata,enctest],axis=1)
```

```python
final_train.head()
```

```python
final_test.head()
```

---
Splitting Data into Features(dataframe) and Target(series):
=
---

```python
x_train=final_train[['age','sleep_hours_per_day','screen_time_hours_per_day','family_history_diabetes','hypertension_history','cardiovascular_history','bmi','waist_to_hip_ratio','systolic_bp','diastolic_bp','heart_rate','cholesterol_total','hdl_cholesterol','ldl_cholesterol','triglycerides','glucose_fasting','glucose_postprandial','insulin_level','hba1c','gender_Female','gender_Male','gender_Other','smoking_status_Current','smoking_status_Former','smoking_status_Never']].copy()
x_test=final_test[['age','sleep_hours_per_day','screen_time_hours_per_day','family_history_diabetes','hypertension_history','cardiovascular_history','bmi','waist_to_hip_ratio','systolic_bp','diastolic_bp','heart_rate','cholesterol_total','hdl_cholesterol','ldl_cholesterol','triglycerides','glucose_fasting','glucose_postprandial','insulin_level','hba1c','gender_Female','gender_Male','gender_Other','smoking_status_Current','smoking_status_Former','smoking_status_Never']].copy()
y_train=final_train['diagnosed_diabetes']
y_test=final_test['diagnosed_diabetes']
```

---
Scaling the data :
=
---

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.set_output(transform='pandas')
train=scaler.fit_transform(x_train)
test=scaler.transform(x_test)
```

---
Training a Model
=
---

`Logistic Regression :`
-

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(solver='saga',max_iter=5000,class_weight='balanced',random_state=42)
model.fit(train,y_train)
```

Checking the `accuracy` of the `Logistic Regression` Model
-

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(f'The accuracy of the testing data is {accuracy_score(y_test,model.predict(test))*100} %')
print(f'The accuracy of the training data is {accuracy_score(y_train,model.predict(train))*100} %')
```

```python
print(f'The confusion matrix for testing data :')
sns.heatmap(confusion_matrix(y_test,model.predict(test)),annot=True,fmt='0.1f',cmap='cividis')
```

```python
print(f'The confusion matrix for training data :')
sns.heatmap(confusion_matrix(y_train,model.predict(train)),annot=True,fmt='0.1f',cmap='coolwarm')
```

```python
print(classification_report(y_test, model.predict(test)))
```

```python
print(classification_report(y_train, model.predict(train)))
```
